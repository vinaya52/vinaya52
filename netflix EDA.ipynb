Missing Value Handling
</div>

```python
netflix = netflix.dropna(subset=['title','description'])
```

```python
cat_cols = ['age_certification', 'imdb_id']

for col in cat_cols:
    netflix[col].fillna('Unknown', inplace=True)


num_cols_mean = ['seasons', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']

for col in num_cols_mean:
    netflix[col].fillna(netflix[col].mean(), inplace=True)
```

Clipping
</div>

```python
for x in col_num:
    Q1 = netflix[x].quantile(.25)
    Q3 = netflix[x].quantile(.75)
    IQR = Q3 - Q1
    lower_limit = Q1-1.5*IQR
    upper_limit = Q3+1.5*IQR
    netflix[x]=netflix[x].clip(lower_limit,upper_limit)
```

Countplots & Distplots
</div>

```python
fig, axes = plt.subplots(2,4, figsize=(18,10))

sns.countplot(x='type', data=netflix, ax=axes[0,0],palette='Set2')
axes[0,0].set_title('Type')
sns.countplot(x='age_certification', data=netflix, ax=axes[0,1],palette='Set3')
axes[0,1].tick_params(axis='x', rotation=90)
axes[0,1].set_title('Age Certification')
sns.countplot(x='release_year', data=netflix, ax=axes[0,2],palette='magma')
axes[0,2].tick_params(axis='x', rotation=90)
axes[0,2].set_title('Release Year')
sns.countplot(x='seasons', data=netflix, ax=axes[0,3],palette='viridis')
axes[0,3].set_title('Seasons')
sns.countplot(
    y=netflix['production_countries']
      .str.split(',')
      .explode()
      .value_counts()
      .head(10)
      .index,
    ax=axes[1,0],palette='coolwarm'
)
axes[1,0].set_title('Top 10 Production Countries')

sns.countplot(
    y=netflix['genres']
      .str.split(',')
      .explode()
      .value_counts()
      .head(10)
      .index,
    ax=axes[1,1],palette='Accent'
)
axes[1,1].set_title('Top 10 Genres')
sns.distplot(netflix['imdb_score'], ax=axes[1,2],color='red')
axes[1,2].set_title('IMDb Score Distribution')
sns.distplot(netflix['runtime'], ax=axes[1,3],color='green')
axes[1,3].set_title('Runtime Distribution')
plt.tight_layout()
plt.show()
```
