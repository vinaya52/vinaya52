Model Building
</h2>

```python
x=df.drop('Stress_Level',axis=1)
y=df['Stress_Level']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

<h2 style="background-color:lightbl<h1 style="background-color:#lightblue;

Linear Regression
</h2>

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

<h2 style="background-color:lightbl<h1 style="background-color:#lightblue;
color:deepblue;padding:10px; border-radius:8px;">
KNeighbors Regressor
</h2>

```python
from sklearn.neighbors import KNeighborsRegressor
metric_k=[]
neighbors=np.arange(3,15)
```

```python
for k in neighbors:
    regressors=KNeighborsRegressor(n_neighbors=k,metric='euclidean')
    regressors.fit(x_train,y_train)
    y_prediction=regressors.predict(x_test)
    r22=r2_score(y_test,y_prediction)
    metric_k.append(r22)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k value')
plt.grid()
plt.show()
```

```python
regressors=KNeighborsRegressor(n_neighbors=12,metric='euclidean')
regressors.fit(x_train,y_train)
```

```python
y_prediction=regressors.predict(x_test)
r221=r2_score(y_test,y_prediction)
r221
```
